近日，Nature发表了一个研究团队开发了一种自主学习的人工突触，它能提高人工神经网络的学习速度。该研究表明：记忆就是神经元间相连接突触的强度分布所表示，而学习可以通过突触的重新配置（即可塑性）而实现。这种能自主学习的人工突触有助于无监督学习的进一步发展。机器之心在这里对该研究进行简要介绍

## 模拟大脑
人工智能的发展和进步很大程度上归功于人类大脑模拟技术的支持。在信息技术领域中，这样的AI系统被称作神经网络。包含于神经网络中的算法可以进行训练，以模仿大脑识别语言和图像的模式。然而，运行一个人工神经网络是非常费时费力的。<br>
来自National Center for Scientific Research(CNRS) 和Evry的研究员开发了一种能直接嵌于芯片的人工突触，称为忆阻器（memristor）。这项发明为有快速和节能需求的智能系统的开发铺平了道路，并且忆阻器可以自主学习。<br>
在人类大脑中，突触负责在神经元之间建立连接。连接强度的强化和学习是由突触的激活而得到提升的。忆阻器以相识的方式工作，通过在两个电极之间包裹一层薄薄的铁电质（可以自发计划）制作而成。在电压脉冲的作用下，忆阻器的电阻将发生改变，正如神经细胞的工作方式。当电阻较小的时候，突触连接更强，反之亦然。忆阻器的学习能力正式源于其电阻可调的性质。
## 更强大的AI
在过去几年内，AI系统得到了令人瞩目的发展，唯一的限制就是学习数据过程中大量的时间和能源消耗，如果能克服这一点，将称为AI进一步发展的跳板。在忆阻器的助力下，学习效率将大大提高，以后的研究工作将延伸到忆阻器上，特别是研究优化函数的方法。作为出发点，研究院成功的建立了一个物理模型以帮助理解忆阻器的工作模式。他们的工作已经发表在Nature Communications上。<br>

## 以下是该研究的概要：
#### 这段看不懂？？？
在大脑中，学习是通过突触重新配置神经元间的连接强度（突触的可塑性）而实现的。在称之为忆阻器的固态突触中，电导率可以通过电压脉冲进行精确的调整，并且根据称为spike-timing-dependent plasticity(STDP)的生物学习规律进行演化。未来神经形态架构包含数十亿这样的纳米突触，这就需要对可塑性的物理机制有一个清晰明确的认识。我们的论文报告基于铁电质隧道结点(ferroelectric tunnel junctions)，并且我们的研究表明STDP能从非均匀极化转化(inhomogeneous polarization switching)中得到处理。通过组合的扫描探针成项、电传输和原子尺度的分子动力学(molecular dynamics)，我们展示了电导变化能通过成核主导(nucleation-dominated）的域逆转进行建模。基于这种无理建模，我们的模拟展示了铁电质纳米突触(ferroelectric nanosynapses)阵列能自主地学习并预测识别模式，这为脉冲神经网络的无监督学习开辟了一条新路径。
#### 简介
皮质信息可以从突触之间的可变连接强度从一个神经元流到另一个神经元。突触强度的总体分布为神经网络提供了记忆，而学习可以通过突触的重新配置（即可塑性）而实现。现已提出了几种调节突触强度演化的机制。特别有潜力的是spike-timing-dependent plasticity(STDP)生物学习规则，它的突触强度会根据来自邻近神经元的时间和因果的电信号而发生改变（如图）。如同在生物系统中观察的那样，STDP能在没有任何外部控制突触强度或任何需要处理信息的先验知识下实现学习。这令STDP可以成为自主无监督学习的基础，因此用人工神经网络实现STDP将成为实现自适应电子架构的重要里程碑。

#### 讨论
我们现在利用这个物理模型在基于铁电质突触的脉冲神经网络中模拟无监督学习。这些模拟将作为测试平台以研究STDP信号波形对网络识别图像模式能力的影响（水平线，对角线和垂直线的条形分别是A，B和C，如图）。模拟网络由一个按9*5横梁(crossbar)排列的铁电质忆阻器阵列构成（a）。9个脉冲输入神经元中每一个都编码了含噪图像的一个像素，其中包含了三种模式之一以进行识别。输出神经元将流经每一列的忆阻器的输入信号积分，然后在达到阈值的时候激活。b图展示了随着图像展示的数量的增加，网络识别率也随着提高。同时也展示了每一行中9个忆阻器中的电导率的变化情况，通过彩色的3*3像素图像表示。在网络学习的时候，第1，2，4的电导率图像最后都汇集为三个输入模式之一，而第3，5行的医嘱器电导率仍然保持随机。经过学习之后，正如c图所说明的，其中的3个输出神经元（1，2，3）汇集了三个输入模式的每一个，并在相关输入出现的时候激活。我们的研究揭示了，成功的无监督学习相当依赖于训练前和训练后的神经元脉冲的确切的信号波形。d图展示了，根据之前所示的例子，在经过精心调控训练后神经元脉振幅在0.82到0.98V之间后，识别率达到100%。然而，当学联后神经元脉冲波形仅仅稍微地狱或者高于那个范围的时候，识别率迅速下降。我们的模拟强调了对忆阻器的动力学的精确只是的了解的重要性，然后才能据此建立精确的物理模型。
![通过铁电质的无监督学习](https://raw.githubusercontent.com/HuangYiran/readPaper/master/fotos/tiedianzhiwujiandu.png)
我们已经确定STDP能从铁电质忆阻器本质上的非均匀计划转换中获得利用。结合实践依赖的传输测量，铁电质成像和搞笑的哈密顿原子MD模型，我们展示了铁电质的转换可以潜在的电阻变化可以通过称述的成核限制模型来描述。通过这种无理模型，我们可以可靠的预测具有不同神经输入的铁电质突触的电导率变化。这些结果为在未来脑启发计算中实现数十亿的可预测人工突触（如深度神经网络）铺平了道路。
